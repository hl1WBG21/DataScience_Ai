```python
landing_outcomes=df.Outcome.value_counts()
#pd.value_counts(df.Outcome)
```

```python
landing_outcomes
```

```python
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome[0])
```

```python
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
bad_outcomes=[]
for i,outcome in enumerate(landing_outcomes.keys()):
    #print(i,landing_outcomes.keys()[i][0])
    if outcome[0] in ["N","F"] :
      #print(outcome)
      bad_outcomes.append(outcome)
```

```python
bad_outcomes
```

```python
landing_classe_0=[]
for i,outcome in enumerate(landing_outcomes.keys()):
    if outcome in bad_outcomes :
      landing_classe_0.append(0)
    else :
      landing_classe_0.append(1)
```

```python
landing_classe_0
```

```python
landing_classe=[]
for i,outcome in enumerate(df.Outcome) :
  if outcome in bad_outcomes :
    landing_classe.append(0)
  else :
    landing_classe.append(1)
```

```python
df["classe"]=landing_classe
```

```python
# the success rate
df["classe"].mean()
```
